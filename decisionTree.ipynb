```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston

dict_data = load_boston()

data_boston = pd.DataFrame(dict_data['data'],columns=dict_data['feature_names'])

y= dict_data['target']

data_boston.head()

DecisionTreeRegressor(criterion='mae', min_samples_split='10',max_depth='5')

model_tree = DecisionTreeRegressor(max_depth=3)
model_tree.fit(data_boston,y)
plt.figure(figsize=(15,10))
tree.plot_tree(model_tree)
plt.show()

feature_importances = pd.Series(model_tree.feature_importances_,index=data_boston.columns)
feature_importances
```
